## 変数と型
Pythonでは変数(もっというと全てのオブジェクト)の型(type)を指定する必要はない。
指定する必要がない⇒便利！ なことばかりではないので注意

### 代表的な型
type()を用いることで変数(オブジェクト)の型を確認することができる(デバッグの際にとても重要！)

```python
# string型(文字列)
sentense = "これは文です。"
print(sentense, type(sentense))

# int型(整数)
a = 100
print(a, type(a))

# float型(浮動小数点数)
b = 50.0
print(b, type(b))

# bool型(真偽値)
c = True
d = False
print(c, type(c))
print(d, type(d))
```

### 変数の型変換
str()関数やint()関数を用いることで、変数の型を強制的に指定することができる

```python
c = 50.24
print(c, type(c))

c_typed_int = int(c)
print(c_typed_int, type(c_typed_int)) # float型をint型にすると切り捨てとなる
# print(c, type(c))


c_typed_str = str(c)
print(c_typed_str, type(c_typed_str)) # str型か数値なのかは(typeで確認しないと)見分けがつかない

d = 'I am Number!'
print(d, type(d))

d_typed_int = int(d) # 文字列(str)は数値に変換できない(エラーが出るはず)

'''
---------------------------------------------------------------------------
ValueError                                Traceback (most recent call last)
<ipython-input-20-f845b0a85b1a> in <module>()
     11 print(d, type(d))
     12 
---> 13 d_typed_int = int(d) # 文字列(str)は数値に変換できない(エラーが出るはず)
     14 print(d_typed_int, type(d_typed_int))

ValueError: invalid literal for int() with base 10: 'I am Number!'
'''
```

練習問題 1 - 30 : 100までの素数のリストを表示せよ（エラトステネスのふるいを使うとよい)

```python
primes = []

for i in range(2, 100):
  isPrime = True
  for p in primes:
    if p*p > i:
      break
    if i % p == 0:
      isPrime = False
      break
  if isPrime:
    primes.append(i)
  
print(primes)
```

## 関数

#### 関数の基本

必要最小限の形

```python
def func(a, b):
  return a + b
```

```python
c = func(10, 20)
print(c)
```

特に関係のない処理(今回の例では`print`)を関数の中に入れることができる。この処理は、関数が呼び出されたときに実行される。

```python
def func2(a, b):
  print(f'calculating {a} + {b}')
  tashizan = a + b # tashizanはローカルな変数なので、関数の外で呼び出すことができない
  return tashizan
```

```python
d = func2(10, 30)
print(d)
```

条件に応じて、処理や返り値を変えることができる。何かしらの値がreturnされると、以降の処理は全て無視される点に注意

```python
def func3(a, b):
  if a%2 == 0:
    print(f'calculating {a//2} + {b}')
    return a//2 + b
  if b%3 == 0:
    print(f'calculating {a} + {b//3}')
    return a + b//3
  if a > 100:
    print(f'{a} is too big, bro.')
```

```python
print(func3(4, 10))
```

```python
print(func3(5, 9))
```

```python
print(func3(4, 9))
```

返り値が何も指定されないまま関数の処理が終了したときは、返り値に`None`値が入る

```python
print(func3(200001, 1))
```

返り値をカンマで区切ることで(つまりタプルとすることで)、複数の値を返すことが可能

```python
def sum_and_product(a, b):
  return a+b, a*b
```

```python
summed, producted = sum_and_product(5, 6) # タプルのアンパックを使っている

print(summed)
print(producted)
```

練習問題 1 - 40 : リストの要素全てを足し上げる関数を作成しなさい

```python
def summed_list(lst):
  summed = 0
  for i in lst:
    summed += i
  return summed
```

```python
test1 = [i+0.5 for i in range(10) if i%2==0]
print(test1)
print(summed_list(test1))

test2 = [i for i in range(51)]
print(test2)
print(summed_list(test2))
```

#### (発展)引数にデフォルト値をもつ関数

```python
def say_something_and_add(x, y, say='something'):
  '''デフォルト値を持つ引数は必ず必須引数の後に定義する'''
  print(say)
  return x + y
```

```python
#デフォルト値は省略できる
print(say_something_and_add(10, 20))
```

```python
#他の値を代入することができる
print(say_something_and_add(20, 10, say='freedom!'))
```

#### (発展)無名関数 lambda

1つの引数と1つの返り値のみで良い関数は、無名関数として一行で書くことができる

```python
# こう書くのではなく
def double_function(x):
  return 2*x

print(double_function(3))

# こう書いて良い
double_function2 = lambda x : 2*x
print(double_function2(3))
```
